```python
import pandas as pd
changed_dataframe = pd.read_csv('final_dataframe.csv')
#the final_dataframe.csv file contains the required data as mentioned in the previous colab notebook
```

```python
changed_dataframe.head()
#view the first few rows of data
```

```python
changed_dataframe.drop(columns = ['index', 'crime count'], inplace = True)
#drop the unwanted columns
```

```python
changed_dataframe.head()
```

```python
changed_dataframe['Occurrence of Crime'].value_counts()
#verify the number of rows representing "yes" and "no"
```

```python
changed_dataframe.info()
#view the dtypes of the attributes
```

```python
changed_dataframe.head()
```

```python
#use label encoder in order to encode the columns of type object or category (Time Range and Occurence of crime)
#We have used label encoder since each column has a fixed number of categories.
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

columns_to_encode = [col for col in changed_dataframe.columns if changed_dataframe[col].dtype == 'object' or changed_dataframe[col].dtype == 'category']
for col in columns_to_encode:
  changed_dataframe[col] = encoder.fit_transform(changed_dataframe[col])
```

```python
from sklearn.model_selection import train_test_split
#from imblearn.over_sampling import RandomOverSampler
#X represents all the columns (independent columns) except occurence of crime.
#y is our dependent variable (Occurence of Crime)
X = changed_dataframe.drop(columns = ['Occurrence of Crime'])
y = changed_dataframe['Occurrence of Crime']
#test size comprises 30% of the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle = True)
```

```python
from sklearn.ensemble import RandomForestClassifier
#use random forest classifier with 100 trees.
model = RandomForestClassifier(random_state=42, n_estimators=100)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

print('Training Accuracy: ', accuracy_score(y_train, y_train_pred)
      ,'Testing Accuracy: ', accuracy_score(y_test, y_test_pred))

#the data is most likely being overfit, so we will use smote later to balance out the datasets
#We obtain a training accuracy of 1 and a testing accuracy of 74.9%
```

```python
feature_importance = pd.DataFrame(list(zip(X_train.columns, model.feature_importances_)))
feature_importance.sort_values(1, ascending=False).head()
#the importance of features according to random forest
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_pred))
#print out the classification report
```

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

#it better handles the overfitting

# Create and train the XGBoost classifier with a standard scaler
model2 = make_pipeline(XGBClassifier(random_state=42))
model2.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model2.predict(X_test)

# Evaluate the model
y_train_pred = model2.predict(X_train)
y_test_pred = model2.predict(X_test)

print('Training Accuracy: ', accuracy_score(y_train, y_train_pred)
      ,'Testing Accuracy: ', accuracy_score(y_test, y_test_pred))

print("Classification Report:")
print(classification_report(y_test, y_test_pred))
#we obtain a training accuracy of 96.68% and a testing accuracy of 72.8%
```

```python
from sklearn.linear_model import LogisticRegression
model3 = LogisticRegression()
model3.fit(X_train, y_train)
#create a logisitc regression model and fit the data
```

```python
y_pred = model3.predict(X_test)

# Evaluate the model
y_train_pred = model3.predict(X_train)
y_test_pred = model3.predict(X_test)

print('Training Accuracy: ', accuracy_score(y_train, y_train_pred)
      ,'Testing Accuracy: ', accuracy_score(y_test, y_test_pred))

print("Classification Report:")
print(classification_report(y_test, y_test_pred, zero_division=0))
#we obtain a training accuracy of 71.4% and a testing accuracy of 72.04%
```

## **AFTER SMOTE**

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
#use smote to balance the dataset (it creates synthetic samples for the minority class (here, yes))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42, n_estimators=100)
model.fit(X_train_smote, y_train_smote)
#retrain random forest on smote modified data
```

```python
from sklearn.metrics import accuracy_score
y_train_pred = model.predict(X_train_smote)
y_test_pred = model.predict(X_test)

print('Training Accuracy: ', accuracy_score(y_train_smote, y_train_pred)
      ,'Testing Accuracy: ', accuracy_score(y_test, y_test_pred))
#training accuracy is 99.9% and testing accuracy is 68.32%
```

```python
feature_importance = pd.DataFrame(list(zip(X_train_smote.columns, model.feature_importances_)))
feature_importance.sort_values(1, ascending=False).head()
#importance of features
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_pred))
#print out the classification report.
```

```python
from xgboost import XGBClassifier
from sklearn.pipeline import make_pipeline

# Create and train the XGBoost classifier with a standard scaler
model2 = make_pipeline(XGBClassifier(random_state=42))
model2.fit(X_train_smote, y_train_smote)

# Make predictions on the test set
y_pred = model2.predict(X_test)

# Evaluate the model
y_train_pred = model2.predict(X_train_smote)
y_test_pred = model2.predict(X_test)

print('Training Accuracy: ', accuracy_score(y_train_smote, y_train_pred)
      ,'Testing Accuracy: ', accuracy_score(y_test, y_test_pred))

print("Classification Report:")
print(classification_report(y_test, y_test_pred))
#we obtain a training accuracy of 94.18% and a testing accuracy of 67.8%
```

```python
from sklearn.linear_model import LogisticRegression
model3 = LogisticRegression()
model3.fit(X_train_smote, y_train_smote)
#create a logisitc regression model and fit the smote modified data
```

```python
y_pred = model3.predict(X_test)

# Evaluate the model
y_train_pred = model3.predict(X_train_smote)
y_test_pred = model3.predict(X_test)

print('Training Accuracy: ', accuracy_score(y_train_smote, y_train_pred)
      ,'Testing Accuracy: ', accuracy_score(y_test, y_test_pred))

print("Classification Report:")
print(classification_report(y_test, y_test_pred, zero_division=0))
#we obtain a training accuracy of 55.27% and a testing accuracy of 55.24%
```
